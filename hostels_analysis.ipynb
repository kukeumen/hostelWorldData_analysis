# 1. Load Data

```python
import pandas as pd
```

각 월별 데이터 불러오기

```python
december_df = pd.read_csv("12월_hostelsWorld_data.csv")
december_df
```

```python
january_df = pd.read_csv("1월_hostelsWorld_data.csv")
january_df
```

```python
february_df = pd.read_csv("2월_hostelsWorld_data.csv")
february_df
```

```python
hostels_df = pd.concat([december_df, january_df, february_df], axis=0)
hostels_df
```

# 2.  Preprocessing

1. room_charge = e 인 데이터 삭제
2. num_reviews 숫자만 남기기
3. num_reveiws = 0인 데이터 삭제
4. review_total_score = No인 데이터 삭제    
(3,4 삭제하는 이유: 리뷰가 없으면 좋고 나쁨을 가리기 어렵기 때문.
5. 주말, 평일정보 추가
6. distance 숫자만 남기기
7. room_charge 숫자만 남기기 (\ 삭제)
8. room_type 숫자로 표시 (Private=0, dorms=1)

```python
# 1. room_charge = e 인 데이터 삭제
# 829개의 데이터 삭제됨.
temp_df = hostels_df[hostels_df['room_charge']!='e']
hostels_df = temp_df
hostels_df = hostels_df.reset_index(drop=True)
hostels_df
```

```python
# 2. num_reviews 에서 문자열 없애고 숫자만 남기기
hostels_df['num_reviews'] = [int(x[0]) for x in hostels_df['num_reviews'].str.split(" ")]
hostels_df
```

```python
# 3. num_reveiws == 0인 데이터 삭제
# 65개 데이터 삭제
temp_df = hostels_df[hostels_df['num_reviews']!=0]
hostels_df = temp_df
hostels_df = hostels_df.reset_index(drop=True)
hostels_df
```

```python
# 4. review_total_score = No인 데이터 삭제
# 256개 데이터 삭제
temp_df = hostels_df[hostels_df['review_total_score']!='NO']
hostels_df = temp_df
hostels_df = hostels_df.reset_index(drop=True)
hostels_df
```

```python
import numpy as np
# 5. 주말, 평일정보 추가 (주말 = 토, 일 입실, 평일 = 월~금 입실)
sundays = ['2022-12-04', '2022-12-11', '2022-12-18', '2022-12-25', '2023-01-01', 
            '2023-01-08', '2023-01-15', '2023-01-22', '2023-01-29',
            '2023-02-05', '2023-02-12', '2023-02-19', '2023-02-26']

hostels_df['weekends']=int(0) #weekends column 생성
#weekends =0: 평일, weekends=1: 주말

# checkout_date == sundays: 토요일 입실, 일요일 퇴실
# checkin_date == sundays: 일요일 입실, 토요일 퇴실

for i in range(len(hostels_df['checkin_date'])):
    if str(str(hostels_df['checkin_date'][i]) in sundays)=='True':
        hostels_df['weekends'][i]=1
        
    if str(str(hostels_df['checkout_date'][i]) in sundays)=='True':
        hostels_df['weekends'][i]=1

# 주말 739개, 평일 2149개
hostels_df
```

```python
#주말 데이터 확인
hostels_df[hostels_df['weekends']==1]
```

```python
# 평일 데이터 확인
hostels_df[hostels_df['weekends']==0]
```

```python
# 6. distance 숫자만 남기기

for i in range(len(hostels_df['distance'])):
    item = hostels_df['distance'][i]
    start_idx = item.index('-')
    end_idx = item.index('km')
    hostels_df['distance'][i] = item[start_idx+1:end_idx].strip()

hostels_df
```

```python
#7. room_charge 숫자만 남기기 (\ 삭제)
for i in range(len(hostels_df['room_charge'])):
    hostels_df['room_charge'][i] = int(hostels_df['room_charge'][i][1:])
hostels_df
```

```python
#8. room_type 숫자로 표시 (Private=0, dorms=1)
for i in range(len(hostels_df['room_type'])):
    if hostels_df['room_type'][i] == 'PRIVATE':
        hostels_df['room_type'][i] = 0
    elif hostels_df['room_type'][i] == 'DORMS':
        hostels_df['room_type'][i] = 1
    else:
        print(i,"is ERROR!")

hostels_df
```

# 3-1) 성수기 숙소의 가격 비교

```python
# 12월, 1월, 2월 가격 비교
# 월별 가격비교를 위해 month_columns 추가
hostels_df['month']=np.nan

for i in range(len(hostels_df['checkin_date'])):
    if ('2022-12' in hostels_df['checkin_date'][i])==True:
        hostels_df['month'][i]="DEC"
    elif ('2023-01' in hostels_df['checkin_date'][i])==True:
        hostels_df['month'][i]="JAN"
    elif ('2023-02' in hostels_df['checkin_date'][i])==True:
        hostels_df['month'][i]="FEB"
    else:
        print(i,"is ERROR!")
        
plt.scatter(hostels_df['month'], hostels_df['room_charge'])
```

```python
# 평일과 주말 가격 비교를 위해 month 컬럼에 월별 평일, 주말 정보 추가
for i in range(len(hostels_df['weekends'])):
    if hostels_df['weekends'][i]==1:
        hostels_df['month'][i] = str(hostels_df['month'][i]+' weekend')
    elif hostels_df['weekends'][i]==0:
        hostels_df['month'][i] = str(hostels_df['month'][i]+' weekday')
    else:
        print(i, "is ERROR!")
```

```python
plt.figure(figsize=(13,6))
plt.scatter(hostels_df['month'], hostels_df['room_charge'])
plt.show()
```

## 성수기 합리적인 숙소 찾기

'합리적인'의 의미
- 거리가 가까울수록
- 가격이 저렴할수록
- 리뷰가 많을수록
- 리뷰 평점이 높을수록   
좋은 숙소라고 설정.

distance, room_charge, num_reviews, review_total_score 가 평균 이상으로 좋은 숙소 추출

```python
# checkin_date가 2022-12-18, 2023-01-08 인 데이터 index 검색
peak_season_hostels_20221218_idx = hostels_df[hostels_df['checkin_date']=="2022-12-18"].index[0]
peak_season_hostels_20230108_idx = hostels_df[hostels_df['checkin_date']=="2023-01-08"].index[-1]

print(peak_season_hostels_20221218_idx)
print(peak_season_hostels_20230108_idx)
```

```python
# checkin_date, checkout_date 가 문자열 데이터이기 때문에 숫
peak_season_hostels_df = hostels_df.iloc[peak_season_hostels_20221218_idx:peak_season_hostels_20230108_idx+1]
peak_season_hostels_df = peak_season_hostels_df.reset_index(drop=True)
peak_season_hostels_df
```

```python
peak_season_hostels_df.describe()
```

```python
# 합리적인 숙소를 간추리기 위해 데이터 가공
# 각 feature 의 평균의 소수점 미만은 절사하여 평균값으로 설정

# distance 가 8.5km 초과인 숙소 삭제
peak_season_hostels_df = peak_season_hostels_df[peak_season_hostels_df['distance']<=8.5]
# room_charge 79,194원 초과인 숙소 삭제
peak_season_hostels_df = peak_season_hostels_df[peak_season_hostels_df['room_charge']<=79194]
# num_reviews 465개 미만인 데이터 삭제
peak_season_hostels_df = peak_season_hostels_df[peak_season_hostels_df['num_reviews']>=465]
# review_total_score 9점 미만 데이터 삭제
peak_season_hostels_df = peak_season_hostels_df[peak_season_hostels_df['review_total_score']>=9]

peak_season_hostels_df = peak_season_hostels_df.reset_index(drop=True)
peak_season_hostels_df
```

```python
peak_season_good_hostels = peak_season_hostels_df['hostel_name'].unique()
print("peak_season_good_hostels: ", peak_season_good_hostels)
print("num of peak_season_good_hostels: ", len(peak_season_good_hostels))
```
